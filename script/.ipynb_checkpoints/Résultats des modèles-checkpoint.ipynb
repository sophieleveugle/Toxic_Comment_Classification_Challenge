```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
res = pd.read_csv("../data/model_result.csv", sep=';')
```

```python
sns.set_palette("Set2")
plt.figure(figsize=(8, 5))
g = sns.scatterplot(y=res['AUC'], x=res["Model"], hue=res['Type'], marker="X", s=150, legend=False)
plt.hlines(y=0.98805, xmin=0, xmax=len(res['AUC']), color="grey", linestyles="dotted")
plt.xlabel(None)
plt.ylabel('Average AUC', fontsize=16)
plt.xticks([])
plt.yticks(fontsize=16)
for line in range(0,res.shape[0]):
     g.text(line+0.1, res.AUC[line], res.Model[line], horizontalalignment='left', size=12, color='black')
plt.savefig('../plots/model_results.png', bbox_inches='tight')
```
